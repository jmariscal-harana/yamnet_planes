```python
## Run shell script in a loop to split individual audios into t second segments
path_in = "/home/catec/train_detector/holdout_noise_sounds/"
path_out = "/home/catec/train_detector/holdout_noise_sounds/2sec/"
t_segments = 2

#get .wav file names in folder without .wav extension
import glob
filenames = glob.glob(path_in+"/*.wav")

#for loop using file names
import os
import subprocess 
i = -1
for filename in filenames:
    #take audio, split it, and save it into a new folder
    i+=1
    filename_no_path = os.path.basename(filename)
    filename_no_path_no_ext, file_extension = os.path.splitext(filename_no_path)
    print("Splitting audio file \""+filename_no_path+"\" into "+str(t_segments)+" second segments")
    bash_command = "ffmpeg -i "+filename+" -f segment -segment_time "+str(t_segments)+" -c copy "+path_out+filename_no_path_no_ext+"_%04d.wav"    
    os.system(bash_command)
```

```python
## Run shell script in a loop to perform a dynamic loudness equalization (dynaudnorm)
def dyn_loudness_equalisation(path_root,path_in,path_out):
    #get .wav file names in folder without .wav extension
    import glob
    filenames = glob.glob(path_in+"*.wav")
    #filenames = [path_in+"track_04_0000.wav"]

    config_norm = "dynaudnorm=g=3:m=50:b=1" #window size=3; max gain factor=50; boundary mode=enabled

    #for loop using file names
    import os
    import subprocess 
    i = -1
    for filename in filenames:
        #take audio, split it, and save it into a new folder
        i+=1
        filename_no_path = os.path.basename(filename)
        filename_no_path_no_ext, file_extension = os.path.splitext(filename_no_path)
        print("Dynamic normalisation of audio file: \""+filename_no_path+"\"")
        bash_command = "ffmpeg -i "+filename+" -af "+config_norm+" -vcodec copy "+path_out+filename_no_path_no_ext+"_norm.wav"    
        os.system(bash_command)

path_root = "/home/ups/Proyectos/Vigia_sonido/audio_ATLAS/"
path_in = path_root+"Split_3/"
path_out = path_root+"Split_3_norm/"

dyn_loudness_equalisation(path_root,path_in,path_out)
```
